```python
from sklearn.model_selection import train_test_split
visitors=pd.read_table('C:/Users/donghyunkim/seoul_visitors.csv',engine='python',sep=',',thousands = ',')
```

```python
visitors.head()
visitors=visitors.dropna()
visitors
```

```python
from matplotlib import cm
#feature
X=visitors[['남성인원','여성인원','승무원인원']]
Y=visitors['label']
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
#parameter change
k_range=range(1,10)
counts=[]

for k in k_range:
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,Y_train)
    counts.append(knn.score(X_test,Y_test))
```

```python
counts
```

```python
lookup_area_name=dict(zip(visitors.label.unique(), visitors.대륙.unique()))
```
